```python
data.groupby('State_name')[['Male','Male_Workers','Female','Female_Workers']].sum()
```

```python
data['Not_working_Males']=data['Male']-data['Male_Workers']
data['Not_working_Males']
```

```python
data.groupby('State_name')[['Male','Male_Workers','Not_working_Males']].sum()[10:-2].plot(kind='bar')
plt.title("Male in each state Vs Male_Workers Vs Not_working_Males ")
plt.show()
```

```python
data['Not_working_Females']=data['Female']-data['Female_Workers']
data['Not_working_Males']
```

```python
data.groupby('State_name')[['Female','Female_Workers','Not_working_Females']].sum()[10:-1].plot(kind='bar')#Female in each state Vs Female_Workers Vs Not_working_Femalesfor each state
plt.title(" Female in each state Vs Female_Workers Vs Not_working_Females ")
plt.show()
```

```python
sns.kdeplot(data['Population'], color='blue',fill=True, shade=True)

plt.title('Population Age Distribution (Density Plot)')
plt.ylabel('Density')
plt.show()
```

```python
data.groupby('State_name')['Population'].sum().sort_values(ascending=False)[:19].plot(kind='bar',color='b')
plt.title("Population for States ")
plt.xlabel("Population")
plt.show()
```

```python
data.groupby('State_name')['Population'].sum().sort_values(ascending=False)[0:7].plot.pie(autopct='%1.1f%%', startangle=90)
```

```python
plt.scatter(data['Secondary_Education'], data['Graduate_Education'],s=40, c='blue', alpha=0.5, edgecolors='black', linewidths=.8)
plt.title("Relation between Secondary_Education Vs Graduate_Education")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
